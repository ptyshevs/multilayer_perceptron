```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-100, 100, 5)

f = lambda x: x ** 2 + 3
Y = f(X) + np.random.randn(X.shape[0]) * 500
```

```python
import keras
from keras.models import Sequential, Model
from keras.layers import Dense, Input
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
```

```python
main_input = Input(shape=(1,), name='main_input')
x = Dense(20, activation='relu')(main_input)
# x2 = Dense(10, activation='relu')
output = Dense(1, activation='linear')(x)

model = Model(inputs=[main_input], outputs=[output])
model.compile(optimizer=Adam(0.1), loss=['mse'])
```

```python
model.fit(X, Y, epochs=500, verbose=True, shuffle=True)
```

```python
pred = model.predict(X)
```

```python
plt.plot(X, Y, c='blue')
plt.plot(X, pred, ls='--', c='red');
```
