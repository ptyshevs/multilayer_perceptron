```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
x = np.linspace(0, 10, num=1100)
f = lambda x: x ** 2
y = f(x)
x_normed = (x - x.mean()) / x.std()
```

```python
model = Sequential()
for i in range(100):
    model.add(Dense(1, activation='sigmoid', use_bias=False))
model.add(Dense(1))
model.compile('adam', loss='mse')

model.fit(x_normed[:, np.newaxis], y[:, np.newaxis], epochs=300, verbose=1)
```

```python
model = Sequential()
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(1))
model.compile('adam', loss='mse')

model.fit(x_normed[:, np.newaxis], y[:, np.newaxis], epochs=300, verbose=1)
```

```python
model.predict(x_normed) * x.std() + x.mean()
```
