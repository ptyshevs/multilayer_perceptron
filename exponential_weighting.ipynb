```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(.1, 15, .3)
```

```python
f = lambda x: np.exp(np.sin(x/200))*3*np.log(x)
```

```python
thetas = f(t)
```

```python
v0 = 0
beta = .9995

thetas_averaged = []
v_prev = v0
for i, ti in enumerate(t):
    vi = beta * v_prev + (1 - beta) * thetas[i]
    thetas_averaged.append(vi)
    v_prev = vi

thetas_corrected = []
v_prev = v0
for i, ti in enumerate(t):
    vi = beta * v_prev + (1 - beta) * thetas[i]
#     vi = vi / (beta + 1)
    thetas_corrected.append(vi / (1 - beta ** (i + 1)))
    v_prev = vi
```

```python
plt.figure(figsize=(16, 9))
plt.plot(t, thetas, label='gt');
plt.plot(t, thetas_averaged, alpha=.9, label='exponential average')
plt.plot(t, thetas_corrected, alpha=.9, label='corrected exponential average');
plt.legend();
```
